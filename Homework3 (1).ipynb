```python
#Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

import math  
 
# Запрашиваем координаты X, Y, Z
x = float(input('X = '))
y = float(input('Y = '))
z = float(input('Z = '))
 
# Вычисляем длину вектора
lvec = math.sqrt(x * x + y * y + z * z)
 
# Выводим длину вектора
print('Length = ', lvec)
```

```python
# Нарисуйте трехмерный график двух любых поверхностей второго порядка.
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X1 = np.arange(-5, 5, 0.5)
X2 = np.arange(-10, 10, 0.5)
Y1 = np.arange(-5, 5, 0.5)
Y2 = np.arange(-10, 10, 0.5)
X1, Y1 = np.meshgrid(X1, Y1)
X2, Y2 = np.meshgrid(X2, Y2)
Z1 = X1**2 + Y1**2
Z2 = np.sqrt(X2**2 - Y2**2)
ax.plot_surface(X1, Y1, Z1)
ax.plot_wireframe(X2, Y2, Z2)
ax.scatter(0, 0, 0, 'z', 50, 'red')
show
```

```python
# Напишите код, который будет переводить полярные координаты в декартовы.
import math
def pol2cart(rho, phi):
    x = rho * math.cos(math.radians(phi))
    y = rho * math.sin(math.radians(phi))
    return(x, y)

rho=float(input("Enter the value of rho:"))
phi=float(input("Enter the value of phi in degrees:"))

print("Для полярных координат x = ", rho, "и y = ", phi, "соответствуют декартовы координаты = ", pol2cart(rho, phi))
```
